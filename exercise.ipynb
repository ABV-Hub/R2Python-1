# Plotting financial data

You need to install the package `pandas-datareader`
Documentation of the package can be found [here](https://pandas-datareader.readthedocs.io/en/latest/index.html)




```python
%matplotlib inline
import pandas_datareader as pdr

tickers = ['AAPL', 'MSFT', 'IBM']

start_date = '2010-01-01'
end_date = '2016-12-31'

panel_data = pdr.DataReader(tickers, 'stooq', start_date, end_date)
panel_data.head()
```

```python
data = panel_data['Close']
data.head()
```

```python
year_MMXIII = data['2013-12-31': '2013-01-01']
year_MMXIII.head(3)
```

## Using of `resample()`

```python
weekly = data.resample(rule='W').mean() # rule='W' stays for weekly
weekly.head()
```

We can now plot the **weekly** data.

```python
fig = weekly.plot(y='AAPL', kind='line', lw=0.75, c='y') 
```

# Plotting - seaborn

```python
import seaborn as sns
sns.set(style="whitegrid")

sns.lineplot(data=data, palette="tab10", linewidth=2.5)
```
